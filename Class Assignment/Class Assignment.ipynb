```python
# Q.12 With function using dictionary
def after_salary_inc(salary, rate):
    try:
        if not isinstance(salary, (int, float)):
            raise TypeError("Salary must be numeric")

        if not isinstance(rate, str):
            raise TypeError("Rating must be a string")

        rate = rate.upper()

        if salary <= 500000:
            inc = {'A': 1.16, 'B': 1.12, 'C': 1.10, 'D': 1.06}
        elif salary <= 1000000:
            inc = {'A': 1.14, 'B': 1.10, 'C': 1.08, 'D': 1.06}
        elif salary <= 1500000:
            inc = {'A': 1.08, 'B': 1.06, 'C': 1.04, 'D': 1.00}
        elif salary <= 2300000:
            inc = {'A': 1.07, 'B': 1.05, 'C': 1.04, 'D': 1.00}
        else:
            raise ValueError("Salary range not defined")

        if rate not in inc:
            raise ValueError("Invalid rating selected")

        if inc[rate] == 1.00:
            return "No increment"

        return salary * inc[rate]

    except TypeError as t:
        return f"Type error: {t}"

    except ValueError as v:
        return f"Value error: {v}"

    except Exception as e:
        return f"Unexpected error: {e}"

    finally:
        print("Execution completed")
```

```python
def fee_calculation(subject, analytics, hotel, food_months, transport):
    try:
        # Type validation
        if not isinstance(subject, str):
            raise TypeError("Subject must be string")

        if not isinstance(analytics, str):
            raise TypeError("Analytics must be string")

        if not isinstance(hotel, str):
            raise TypeError("Hotel must be string")

        if not isinstance(transport, str):
            raise TypeError("Transport must be string")

        if not isinstance(food_months, int):
            raise TypeError("Months must be integer")

        subject = subject.upper()
        analytics = analytics.upper()
        hotel = hotel.upper()
        transport = transport.upper()

        fees = 200000

        # Subject validation
        if subject not in ('HR', 'FINANCE', 'MARKETING', 'DS'):
            raise ValueError("Invalid subject selected")

        # Analytics fee
        if analytics == 'Y':
            if subject != 'DS':
                fees += fees * 0.10
            else:
                return "Analytics not available for DS"

        # Hostel fee
        if hotel == 'Y':
            fees += 200000

        # Food fee
        if food_months > 0:
            fees += food_months * 2000
        else:
            raise ValueError("Invalid food month input")

        # Transportation fee
        if transport == 'SEMESTER':
            fees += 13000
        elif transport == 'ANNUAL':
            fees += 26000
        else:
            raise ValueError("Invalid transportation option")

        return f"Total Annual Cost: ₹{int(fees)}"

    except TypeError as t:
        return f"Type error: {t}"
    except ValueError as v:
        return f"Value error: {v}"
    except Exception as e:
        return f"Unexpected error: {e}"
```

```python
class studentinfo:

    def __init__(self, subject, analytics, hostel, food_months, transport):
        self.base_fee = 200000
        self.subject = subject.upper()
        self.analytics = analytics.upper()
        self.hostel = hostel.upper()
        self.food_months = food_months
        self.transport = transport.lower()
        self.total = 0

    
    def course_fee(self):
        if self.subject not in ["HR", "FINANCE", "MARKETING", "DS"]:
            raise ValueError("Invalid Subject Selected")

        fee = self.base_fee

        if self.analytics == "Y":
            if self.subject == "HR" or self.subject == "MARKETING" or self.subject=="FINANCE":
                fee += 0.10 * self.base_fee
            elif self.subject == "DS":
                print("Analytics not available for DS")
        elif self.analytics=="N":
                fee=self.base_fee

        return fee

   
    def hostel_fee(self):
        if self.hostel not in ["Y", "N"]:
            raise ValueError("Hostel option must be Y or N")

        if self.hostel == "Y":
            return 200000
        elif self.hostel=="N":
            return 0

   
    def food_fee(self):
        if self.food_months < 0:
            raise ValueError("Food months cannot be negative")
        return self.food_months * 2000

    
    def transport_fee(self):
        if self.transport == "semester":
            return 13000
        elif self.transport == "annual":
            return 26000
        else:
            raise ValueError("Transport must be semester or annual")

   
    def calculate_total(self):
        self.total = (self.course_fee() + self.hostel_fee() + self.food_fee() + self.transport_fee())
        return self.total

try:
    subject = input("Enter Subject (HR/Finance/Marketing/DS): ")
    analytics = input("Analytics? (Y/N): ")
    hostel = input("Hostel? (Y/N): ")
    food_months = int(input("Food months: "))   
    transport = input("Transportation (semester/annual): ")
    student=studentinfo(subject, analytics, hostel, food_months, transport)
    

    total = student.calculate_total()
    print("Total Annual Cost: Rs", total)

except ValueError as e:
    print("Error:", e)

except Exception as e:
    print("Something went wrong:", e)
```

```python
from datetime import datetime

def calculate_age(dob):
    try:
        # dob format: YYYY-MM-DD
        birth_date = datetime.strptime(dob, "%Y-%m-%d").date()
        today = datetime.today().date()

        age = today.year - birth_date.year

        if (today.month, today.day) < (birth_date.month, birth_date.day):
            age -= 1

        return age

    except ValueError:
        return "Invalid date format! Use YYYY-MM-DD"
```

```python
from datetime import datetime

def check_age_eligibility(dob, eligibility_age):
    try:
        birth_date = datetime.strptime(dob, "%Y-%m-%d").date()
        today = datetime.today().date()

        age = today.year - birth_date.year
        if (today.month, today.day) < (birth_date.month, birth_date.day):
            age -= 1

        return "Eligible" if age >= eligibility_age else "Not Eligible"

    except ValueError:
        return "Invalid DOB format! Use YYYY-MM-DD"
```

```python
import re
import os

file="C:\\Users\\shind\\Desktop\\Just copy\\bill.txt"

def name_():
    name=input('enter name:')
    if not name.replace(' ', '').isalpha():
        raise ValueError('Invalid name added')
    return name

def mobile_():
    mobile=input('enter mobile no:')
    if not (mobile.isdigit() and len(mobile) == 10):
        raise ValueError('Mobile no must be int and 10 digit')
    return mobile

def pan_():
    pan=input('Enter your pan no:').upper()
    if not re.fullmatch('[A-Z]{5}[0-9]{4}[A-Z]', pan):
        raise ValueError('Pan must be in valid format')
    return pan

def tenth_marks_():
    tenth_marks=input('Enter 10 th marks:')
    if not (tenth_marks.isdigit() and 0 <= int(tenth_marks) <= 100):
        raise ValueError('Marks should be in integer')
    return tenth_marks

def twelth_marks_():
    twelth_marks=input('enter 12th marks:')
    if not (twelth_marks.isdigit() and 0 <= int(twelth_marks) <= 100):
        raise ValueError('marks is in int')
    return twelth_marks

def grad_():
    grad=input('enter graduation marks:')
    if not (grad.isdigit() and 0 <= int(grad) <= 100):
        raise ValueError('marks is in int')
    return grad

def records():
    try:
        if os.path.exists(file):
            with open(file, 'r') as f:
                lines=f.readlines()
                emp_id=len(lines) + 1
        else:
            lines=[]
            emp_id=1

        
        pan=pan_()

       
        for line in lines:
            if pan in line:
                return "This pan already exists in the system"

        # Now take remaining inputs
        emp_name=name_()
        emp_mobile=mobile_()
        emp_ten=tenth_marks_()
        emp_twelth=twelth_marks_()
        emp_grad=grad_()

        data={
            'ID': emp_id,
            'Name': emp_name,
            'Mobile': emp_mobile,
            'Pan no': pan,
            '10th': emp_ten,
            '12th': emp_twelth,
            'Graduation': emp_grad
        }

        with open(file, 'a') as f:
            f.write(f'{data}\n')

        print('Records added')
        print("Employee id", emp_id)
        return data

    except Exception as e:
        return f"Error: {e}"
```

```python
menu={'chicken fry':{'Price':120,'category':'Starter'},
      'soup':{'Price':100,'category':'Starter'},
      'chicken masala':{'Price':250,'category':'Main Course'},
      'paneer masala':{'Price':250,'category':'Main Course'},
      'ras malai':{'Price':75,'category':'Dessert'}
	  }

class order():
    def __init__(s,table_no):
        s.table_no=table_no
        s.items=[]             # items
        s.delivery_={}
    def add_menu(s,item_name):
        if item_name in menu:
            s.items.append(item_name)
            s.delivery_[item_name]=False
            print(f"{item_name} is added in menu")
        else:
            print('selected item is not in the menu')
    def delivered(s,item_name):
        if item_name in s.delivery_:
            s.delivery_[item_name]=True
            print(f'{item_name} is delivered')
        else:
            print(f'{item_name} is not delivered')
    def total_bill(s):
        total=0
        for item in s.items:
             total += menu[item]['Price']
        return total
    def bill_print(s):
        print('\n'+'-'*50)
        print(' '*15+"Dipti's Kitchen"+' '*20)
        print('-'*50)
        print(f"Table num: {s.table_no}")
        print('-'*50)
        for item in s.items:
            price=menu[item]['Price']
            status='delivered' if s.delivery_[item] else 'pending'
            print(f"{item:25} {price:<5} {status}")
        print('-'*50)
        print(f"Total Bill: {' '*13} {s.total_bill()}")
        print('-'*50)
               
    def save_bill(s):
        with open("C:\\Users\\shind\\Desktop\\Backup\\bill123.txt",'w') as f:
            f.write(f"'\n'+'-'*50")
            f.write(f"' '*15+'Dipti Kitchen'+' '*20")
            f.write(f"'-'*50")
            f.write(f"Table num: {s.table_no}")
            f.write(f"'-'*50")
            for item in s.items:
                 price=menu[item]['Price']
                 status='delivered' if s.delivery_[item] else 'pending'
                 f.write(f"{item:25} {price:<5} {status}\n")
            f.write(f"'-'*50")
            f.write(f"Total Bill: {' '*13} {s.total_bill()}\n")
            f.write(f"'-'*50")
        print('Bill saves successfuly')
                 
for item,details in menu.items():
	print(f"{item:25} {details['Price']} ({details['category']})") 
table_no= int(input('Enter table no:'))  
order_=order(table_no)

while True:
    item=input('Enter item name (type ok if you are done):').lower()
    if item.lower()=='ok':
        break
    order_.add_menu(item)
while True:
    item=input('Enter delivered item name(or type ok if not):')
    if item.lower()=='ok':
         break
    order_.delivered(item)
order_.bill_print()
order_.save_bill()

```
